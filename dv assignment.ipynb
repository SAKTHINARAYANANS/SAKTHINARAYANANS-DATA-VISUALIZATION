```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("India Agriculture Crop Production.csv")
data.head()
```

```python


crop_production = data.groupby('Crop')['Production'].sum().reset_index()

# Sort the data by production values (optional)
crop_production = crop_production.sort_values(by='Production', ascending=False)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(crop_production['Crop'], crop_production['Production'], color='skyblue')
plt.xlabel('Crop')
plt.ylabel('Production (in tons)')
plt.title('Crop Production in India')
plt.xticks(rotation=90)  # Rotate x-labels for better readability
plt.tight_layout()  # Adjust layout to prevent cropping of labels
plt.show()
```

```python

# Sort the data by production values (optional)
crop_production = crop_production.sort_values(by='Production', ascending=False)

# Plotting the pie chart
plt.figure(figsize=(10, 8))
plt.pie(crop_production['Production'], labels=crop_production['Crop'], autopct='%1.1f%%', startangle=140)
plt.title('Crop Production Distribution in India')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python

crop_production = data.pivot_table(index='Year', columns='Crop', values='Production', aggfunc='sum')

# Plotting the line chart
plt.figure(figsize=(12, 8))
crop_production.plot(kind='line', marker='o', linewidth=2)
plt.title('Crop Production Trends in India')
plt.xlabel('Year')
plt.ylabel('Production (in tons)')
plt.grid(True)
plt.legend(title='Crop', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python

crop_production = data['Production']

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(crop_production, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Production (in tons)')
plt.ylabel('Frequency')
plt.title('Distribution of Crop Production in India')
plt.grid(True)
plt.show()
```

```python

crop_production = data[['Crop', 'Production']]

# Plotting the boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(crop_production['Production'])
plt.xlabel('Crop')
plt.ylabel('Production (in tons)')
plt.title('Crop Production Distribution in India')
plt.xticks([1], ['All Crops'])  # Displaying x-tick as 'All Crops'
plt.grid(True)
plt.show()
```
